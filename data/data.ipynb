```python
import pandas as pd
import numpy as np
```

```python
from datetime import date, timedelta
import time
from datetime import date

dates = []

startyear = "2017"
startmonth = "01"
startday = "31"
endyear = "2019"
endmonth = "03"
endday = "01"
startdate = date(int(startyear), int(startmonth), int(startday))
while startdate < date(int(endyear), int(endmonth), int(endday)):
    dates.append(startdate)
    startdate += timedelta(days=1)
dates_array = np.asarray(dates)
```

```python
df = pd.DataFrame(columns=['date','1', '2', '3', '4', '5', '6', 'total'])
```

```python
df["date"] = dates_array
```

```python
print(df['date'])
```

```python
df['1'] = np.random.randint(1, 21, df.shape[0])
df['2'] = np.random.randint(1, 21, df.shape[0])
df['3'] = np.random.randint(1, 21, df.shape[0])
df['4'] = np.random.randint(1, 21, df.shape[0])
df['5'] = np.random.randint(1, 21, df.shape[0])
df['6'] = np.random.randint(1, 21, df.shape[0])
```

```python
df['total']=df['1']+df['2']+df['3']+df['4']+df['5']+df['6']
```

```python
export_csv = df.to_csv ('daily_data.csv', index = None, header=True)
```
